c.Finally, extract the companies, locations, and posting dates for each job.
For example, the first job has a company of "Payne, Roberts and Davis", a location of "Stewartbury, AA",
and a posting date of "2021-04-08". Ensure that the text that you extract is clean, 
meaning no extra spaces or other characters at the beginning or end. 

```python
title_list = []
company_list = []
location_list = []
post_date_list = []
for title in titles:
    title_element = title.find("h2", class_="title")
    company_element = title.find("h3", class_="company")
    location_element = title.find("p", class_="location")
    timestamp = title.find('time', datetime_='')
    title_list.append(title_element.text.strip())
    company_list.append(company_element.text.strip())
    location_list.append(location_element.text.strip())
    post_date_list.append(timestamp.text.strip())
#     print(title_list)
#     print(company_list)
#     print(location_list)
#     print(post_date_list
      
#   n some code from this website "Real Python" 
```

d.Take the lists that you have created and combine them into a pandas DataFrame. 
    
    


```python
df = pd.DataFrame(list(zip(title_list, company_list, location_list, post_date_list)), columns=['job_title', 'company','location','posting_date'])
```
